```python
import pandas as pd
```

```python
df_less = pd.read_csv('Lokomotiva.csv')
df_less.head()
```

```python
df_osijek = df_less[df_less['label/0/text'] == 'NK Osijek']
df_osijek.head(50)
```

```python
df = pd.DataFrame()

def get_reciever(df, idx, sender):
    while True:
        if idx + 1 >= len(df):
            break
        reciever = df['code'].values[idx]
        if reciever == sender:
            idx = idx + 1
            continue
        else:
            return reciever

for idx, row in df_osijek.iterrows():
    index = idx
    sender = row['code']
    event = row['label/1/text'] ##pass accurate
    pos_x =  row['pos_x']
    pos_y =  row['pos_y']
    start =  row['start']
    end =  row['end']
    
    if event == 'Passes accurate':
        reciever = get_reciever(df_osijek, idx, sender)
        tmp = pd.DataFrame({
            "sender":[sender], 
            "reciever": [reciever],
            "event":[event],
            "pos_x": [pos_x],
            "pos_y": [pos_y],
            "start": [start]
            "end": [end]

        })   
        df = df.append(tmp)
```
